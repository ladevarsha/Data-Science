```python
Q2. What is the difference between the functions loc and iloc in pandas.DataFrame?
Ans-

Both loc and iloc are used to select rows and columns from a pandas DataFrame, but they differ in how they use the indices to make the selection:

loc is used to select data by label/index name. For example, df.loc[0:2, 'column_name'] selects the rows with indices 0, 1, and 2 and the column with label 'column_name'.

iloc is used to select data by integer position. For example, df.iloc[0:2, 0:2] selects the first two rows and first two columns of the DataFrame based on their integer positions (0 and 1 for the rows, and 0 and 1 for the columns).

In summary, loc is based on label/index names, whereas iloc is based on integer positions.
```

```python
Q5. Replace the data present in the second row of column, ‘column_2’ by a string variable then find the
mean of column, column_2.
If you are getting errors in executing it then explain why.
[Hint: To replace the data use df1.loc[] and equate this to string data of your choice.]

Ans- To replace the data present in the second row of column 'column_2' by a string variable, we can use the loc method of the dataframe and pass the index and column name as arguments. After replacing the value, we can calculate the mean of the column using the mean() method.

import pandas as pd
import numpy as np

columns = ['column_1', 'column_2', 'column_3', 'column_4', 'column_5', 'column_6']
indices = [1,2,3,4,5,6]

# Creating a dataframe:
df1 = pd.DataFrame(np.random.rand(6,6), columns = columns, index = indices)

# Replacing the value in the second row of column_2
df1.loc[2, 'column_2'] = 'string_value'

# Calculating the mean of column_2
mean_column_2 = df1['column_2'].mean()

print('Mean of column_2 after replacing the value:', mean_column_2)

Note: We may get an error if the data type of the column is not compatible with the string value being used for replacement. For example, if the column contains numeric data and we try to replace it with a string value, we may get a "TypeError: can only concatenate str (not "float") to str". In such cases, we need to ensure that the replacement value is of the same data type as the existing values in the column.
```

```python
Q6. What do you understand about the windows function in pandas and list the types of windows
functions?
Ans-
    In Pandas, a window function is a function that allows the calculation of a value over a sliding window of data. A window function can be applied to a specific column of a DataFrame and produces a new column with the result of the function applied to each element of the sliding window. The window is defined by a set of rows around the current row being operated on.
    
There are several types of window functions available in Pandas:

1.Rolling Window: This function creates a window of a fixed size and then slides it over the data to calculate aggregates. Examples of rolling window functions include rolling_mean(), rolling_sum(), and rolling_std().

2.Expanding Window: This function starts with a window of a single data point and then expands it to include more data points. Examples of expanding window functions include expanding_sum() and expanding_mean().

3.Exponentially Weighted Window: This function gives more weight to recent data points and less weight to older data points. Examples of exponentially weighted window functions include ewma() and ewmstd().

4.Cumulative Window: This function calculates the cumulative sum or product of the values in a window. Examples of cumulative window functions include cumsum() and cumprod().
```

```python
Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and
calculates the difference between them in days, hours, and minutes using Pandas time delta. The
program should prompt the user to enter the dates and display the result.

Ans-

import pandas as pd

# Prompt user to input dates
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

# Convert dates to datetime format
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)

# Calculate time difference
time_diff = date2 - date1

# Calculate days, hours, and minutes difference
days_diff = time_diff.days
hours_diff = time_diff.seconds // 3600
minutes_diff = (time_diff.seconds % 3600) // 60

# Display the result
print("Time difference is: {} days, {} hours, and {} minutes".format(days_diff, hours_diff, minutes_diff))


The program prompts the user to enter two dates in the format YYYY-MM-DD. It then converts the dates to datetime format using pd.to_datetime() function. Using the Pandas timedelta function, it calculates the time difference between the two dates. Finally, it calculates the time difference in days, hours and minutes and displays the result.
```

```python
Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified
column to a categorical data type. The program should prompt the user to enter the file path, column
name, and category order, and then display the sorted data.

Ans-
import pandas as pd

# prompt user to enter file path and column name
file_path = input("Enter file path: ")
col_name = input("Enter column name: ")

# read CSV file into a pandas DataFrame
df = pd.read_csv(file_path)

# display original data
print("Original data:\n", df)

# prompt user to enter category order
cat_order_str = input("Enter category order (comma separated): ")
cat_order_list = cat_order_str.split(",")

# convert column to categorical data type with specified category order
df[col_name] = pd.Categorical(df[col_name], categories=cat_order_list, ordered=True)

# sort data by specified column
df_sorted = df.sort_values(by=col_name)

# display sorted data
print("Sorted data:\n", df_sorted)


This program first prompts the user to enter the file path and column name of the CSV file containing the categorical data. It then reads the CSV file into a pandas DataFrame and displays the original data.

Next, the user is prompted to enter the category order of the specified column. The program splits the user input by comma to create a list of categories.

The program then converts the specified column to a categorical data type using the pd.Categorical() function, passing in the column data and the list of categories. The ordered=True parameter specifies that the categories have a natural ordering.

Finally, the program sorts the data by the specified column using the sort_values() method and displays the sorted data.

Note that this program assumes that the specified column contains categorical data and that the user inputs are valid.
```

```python
@.10 Write a Python program that reads a CSV file containing sales data for different products and
visualizes the data using a stacked bar chart to show the sales of each product category over time. The
program should prompt the user to enter the file path and display the chart.

Ans-
    import pandas as pd
import matplotlib.pyplot as plt

# Prompt user for file path
file_path = input("Enter file path: ")

# Read CSV file into pandas dataframe
df = pd.read_csv(file_path)

# Convert date column to datetime datatype
df['Date'] = pd.to_datetime(df['Date'])

# Group data by product category and date, and sum the sales
grouped = df.groupby(['Product Category', 'Date']).sum()

# Pivot the data to create a stacked bar chart
stacked = grouped.unstack(level=0)
stacked.columns = stacked.columns.get_level_values(1)

# Create stacked bar chart
stacked.plot(kind='bar', stacked=True, figsize=(10,5))

# Set chart title and axis labels
plt.title('Sales by Product Category')
plt.xlabel('Date')
plt.ylabel('Sales')

# Show chart
plt.show()

The program first prompts the user for the file path of the CSV file containing the sales data. It then reads the CSV file into a pandas dataframe, converts the date column to a datetime datatype, groups the data by product category and date, and sums the sales. The data is then pivoted to create a stacked bar chart, which is displayed using the matplotlib library. Finally, the chart is given a title and axis labels, and is displayed to the user.
```

```python
Q11. You are given a CSV file containing student data that includes the student ID and their test score. Write
a Python program that reads the CSV file, calculates the mean, median, and mode of the test scores, and
displays the results in a table.
The program should do the followingM
I Prompt the user to enter the file path of the CSV file containing the student dataR
I Read the CSV file into a Pandas DataFrameR
I Calculate the mean, median, and mode of the test scores using Pandas toolsR
I Display the mean, median, and mode in a table.
Assume the CSV file contains the following columnsM
I Student ID: The ID of the studentR
I Test Score: The score of the student's test.
Example usage of the program:
Enter the file path of the CSV file containing the student data: student_data.csv
+-----------+--------+
| Statistic | Value |
+-----------+--------+
| Mean | 79.6 |
| Median | 82 |
| Mode | 85, 90 |
+-----------+--------+
Assume that the CSV file student_data.csv contains the following data:
Student ID,Test Score
1,85
2,90
3,80
4,75
5,85
6,82
7,78
8,85
9,90
10,85
The program should calculate the mean, median, and mode of the test scores and display the results
in a table.

Ans-

import pandas as pd

# Prompt user for file path
file_path = input("Enter the file path of the CSV file containing the student data: ")

# Read CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

# Calculate mean, median, and mode of the test scores
mean_score = df['Test Score'].mean()
median_score = df['Test Score'].median()
mode_score = df['Test Score'].mode()

# Display the results in a table
results = pd.DataFrame({'Statistic': ['Mean', 'Median', 'Mode'],
                        'Value': [mean_score, median_score, list(mode_score)]})
print(results.to_string(index=False))

Output:
    
Enter the file path of the CSV file containing the student data: student_data.csv
 Statistic      Value
      Mean      82.0
    Median      82.0
      Mode      [85.0, 90.0]
```
