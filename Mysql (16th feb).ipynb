```python
4.What is DQL?Explain SELECT with an example.
Ans-
    DQL stand for Data Query Language,which is subset of SQL used to retrieve data from databases.DQL statements are used to select and retrieve data from database tables.
    
    The most common DQL statements is SELECT.
       SELECT * from employees;
        
        This statements selects all columns from the "employees" table and returns all the rows in the table.The asterisk(*) is a wildcard that represents all columns in the table.
        
        SELECT id, name ,salary FROM employees WHERE salary > 50000;
        
        this statements selects the "id","name", and "salary"columns from the "employees"table and returns only the rows where the "salary" column is greater than 50000.
```

```python
5.Explain primary key and Foreign key.
Ans-
    A primary key is a unquie identifier for a record or row in a database table.It is used to ensure that each row in a table is unquiely identifiable and can be accesses or updated easily.
    A primary key is a column or set of columns that are selected to identify each row in a table.It must be unique, meaning that no two rows can have the same primary key,and it must not contain null values.
    
    A foreign key,is a field in a table that is used to reference the primary key of another table.It is used to establish a relationship between two tables,where the foreign key in one table refers to the primary key in another table.
    The foreign key coloumn must have the same data type and size as the primary key column it references,its values must exist in the primary key column of the other table.
```

```python
6.Write a python code to connect MySQL to python.Explain the cursor()and execute()method.
Ans -
      To connect MySQL to python,We need to use a python MySQL connector.The most common python MySQL connector is the 'mysql-connector-python'module,which can be installed using pip.
    e.g-   
          import mysql.connector
        
        # establish connection to MySQL
        
        mydb =mysql.connector.connect(
            host = "localhost",
            user='username",
            password="password",
            database="database_name"
        )
        
        #create a cursor object
        
        mycursor = mydb.cursor()
        
        #execute a SQL query
        mycursor.execute("SELECT * from table_name")
        
        #fetch all rows from the query result
        myresult = mycursor.fetchall()
        
        #print the result
        for row in myresult:
            print(row)
            
we create a cursor object using the 'cursor()'method of the connection object.The cursor object is used to execute SQL queries and fetch data from the database.
The 'execute()'method of the cursor object is used to execute a SQL query.We pass the SQL query as a parameter to  the 'execute()'method.
```

```python
7.Give the order of execution of SQL clauses in a SQL query
Ans-
     1.'FROM'clause: This clause specifies the table(s) from which data is retrieved.
    
     2.'JOIN'clause: If there are any joins in the query,they are executed next.This clause specifies how tables are joined together.
        
    3.'WHERE'clause: This clause is used to filter the data based on specified conditions.
    
    4.'GROUP BY" clause: This clause is used to group the data based on specified columns.
    
    5.'HAVING' clause: This clause is used to groups created by the 'GROUP BY'clause based on specified conditions.
    
    6.'SELECT'clause : This clause is used to select the columns that will be included in the result set.
    
    7.'DISTINCT'clause: If the query includes the 'DISTINCT'keyword,duplicates in the result set are removed at this stage.
    
    8.'ORDER BY' clause: This clause is used to sort the result set based on specified columns.
    
    9.'LIMIT' and 'OFFSET' clauses: If specified ,these clauses are used to limit the number of rows returned by the query and skip a specified number of rows,respectively.
```
